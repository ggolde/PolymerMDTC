```python
# vars
monomer_file = 'input/dma_opt.sdf'
box_script_template = 'input/aligned_box_template_21step.lammps'
n_chains = 9
dp_n = 100
n_by_n = [3,3]
```

# Site Control

```python
# generate chain end single chains
# error in 0.5 and beyond
chain_end_dir = 'md/site_control'
tacticities = [0.5]
sub_directories = ['tact' + str(int(p_tact*100)) for p_tact in tacticities]

for tact, sub_dir in zip(tacticities, sub_directories):
    for trial_dir in ['trial3']:
        work_dir = chain_end_dir + '/' + sub_dir + '/' + trial_dir + '/' + 'single_chains'
        polymer_sim.generate_single_chains(monomer_file, n_chains, dp_n, p_tact=tact, log='all_single_chains.log', work_dir=work_dir, chain_end_control=False)
```

```python
tacticities = [0.2, 0.4, 0.5, 0.6, 0.8, 0.9]
sub_directories = ['tact' + str(int(p_tact*100)) for p_tact in tacticities]
tact_strs = [str(int(p_tact*100)) for p_tact in tacticities]
chain_end_dir = 'md/site_control'

for sub_dir, tact_str in zip(sub_directories, tact_strs):
    for trial_dir in ['trial1', 'trial2', 'trial3']:
        work_dir = chain_end_dir + '/' + sub_dir + '/' + trial_dir + '/'
        aligned_box_flag_dict = {
            'LOG':'md.log',
            'SPRING_STRENGTH':'10.0',
            'MAX_PRESS':'5000.0',
            'RESTART':'md.rst',
            'DUMP_MIN':'dump.min.lammpstrj',
            'DUMP_MD':'dump.md.lammpstrj',
            'DUMP_MD_UNWRAPPED':'dump.md_unwrapped.lammpstrj',
            'DUMP_LAST':'dump.final_state.lammpstrj',
            'DATA_OUTPUT':'final.data'
        }
        output_data_name = 'aligned_box_tact' + tact_str
        polymer_sim.make_aligned_box_and_script(monomer_file, tempalte_file=box_script_template, path_to_single_chains=(work_dir + 'single_chains/'), n_chains=n_chains, n_by_n=n_by_n,
                                                output_data_name=output_data_name, work_dir=work_dir, script_flag_dict=aligned_box_flag_dict, chain_names='single_chain')
```
